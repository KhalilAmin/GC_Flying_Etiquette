```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('flying-etiquette.csv')
df.head()
```

```python
# Remove any blanks

non_null_df = df.dropna()
non_null_df.head()
```

```python
# Remove duplicates

no_dup_df = non_null_df.drop_duplicates()
no_dup_df.head()
```

```python
#Change “Under 5 ft, to 5 ft”

no_dup_df['How tall are you?'].replace('Under 5 ft.', "5'0\"")
no_dup_df.head(20)
```

```python
### Create a new column “Height (Inches)” that converts the “How tall are you?”
### column in feet and inches to an integer total of inches.

height_inches_df = no_dup_df.copy()

height_inches_df.insert(4, "Height (Inches)", 0)

def conversion(height):
    try:
        parts = height.split("'")
        feet = int(parts[0])
        inches = int(parts[1][:-1])
        return feet * 12 + inches
    except:
        return None

height_inches_df['Height (Inches)'] = height_inches_df["How tall are you?"].apply(conversion)



height_inches_df.head()
```
